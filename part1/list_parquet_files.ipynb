```python
import glob
import os

import pandas
```

```python
parquet_files = glob.glob('../../data/physionet/mimic3wdb/matched/*/*/*.parquet')
parquet_files = list(filter(lambda f: not(f.endswith('.features.parquet')), parquet_files))
print(len(parquet_files))
parquet_files[:10]
```

```python
df = pandas.DataFrame()
df['file'] = parquet_files
df['size'] = list(map(lambda parquet_file: os.path.getsize(parquet_file), parquet_files))
df['bucket_id'] = list(map(lambda parquet_file: parquet_file.split(os.path.sep)[-3], parquet_files))
df['subject_id'] = list(map(lambda parquet_file: parquet_file.split(os.path.sep)[-2], parquet_files))
df['record_id'] = list(map(lambda parquet_file: '-'.join(parquet_file.split(os.path.sep)[-3:])[:-13], parquet_files))
df.head()
```

```python
df.to_csv('parquet_files.csv', index=False)
```

```python
df[['bucket_id', 'size']].groupby('bucket_id').sum().sort_values(by='size', ascending=False)
```

```python
df[['bucket_id', 'subject_id', 'size']]      \
    .groupby(by=['bucket_id', 'subject_id']) \
    .sum()                                   \
    .sort_values(by='size', ascending=False) \
    .head(10)
```

```python
df[['bucket_id', 'subject_id', 'record_id', 'file', 'size']]      \
    .groupby(by=['bucket_id', 'subject_id', 'record_id', 'file']) \
    .sum()                                                \
    .sort_values(by='size', ascending=False)              \
    .head(10)
```
